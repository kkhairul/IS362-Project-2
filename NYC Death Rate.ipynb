# Q1. Which year had the most deaths ?

```python
deaths.groupby('Year').count()
```

# Answer Q1. We can see 2011 had the most deaths with a total of 43.

# Q2. Which sex had the highest death amount ?

```python
deaths.groupby('Race Ethnicity')['Sex'].count()
```

# Q3. What were the top 5 leading causes of deaths in 2007 ?

```python

deaths[deaths['Year']==2007].sort_values('Deaths', ascending = False).head(5)
```
